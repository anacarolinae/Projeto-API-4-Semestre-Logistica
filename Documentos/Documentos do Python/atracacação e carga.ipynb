```python
import pandas as pd

from google.colab import drive

drive.mount('/content/drive')
```

```python
df_Carga = '/content/drive/My Drive/API/ANTAQ/Dados Bi/Carga.csv'

df_Carga = pd.read_csv('/content/drive/My Drive/API/ANTAQ/Dados Bi/Carga.csv', low_memory=False, sep=';', encoding='UTF-8')
```

```python
df_Carga.info()
```

```python
colunas_desejadas = ['IDCarga', 'IDAtracacao', 'Origem', 'Destino', 'VLPesoCargaBruta']
```

```python
df_Carga_Filtrado = df_Carga[colunas_desejadas]
```

```python
df_Carga_Filtrado.head(5)
```

```python
df_atracacao = '/content/drive/My Drive/API/ANTAQ/Dados Bi/Atracacao_Limpa.csv'

df_atracacao = pd.read_csv('/content/drive/My Drive/API/ANTAQ/Dados Bi/Atracacao_Limpa.csv', low_memory=False, sep=';', encoding='UTF-8')
```

```python
df_atracacao = df_atracacao.merge(
    df_Carga_Filtrado[['IDAtracacao', 'IDCarga']],
    on='IDAtracacao',
    how='left'
)
```

```python
df_atracacao = df_atracacao.merge(
    df_Carga_Filtrado[['IDCarga', 'Destino']],
    on='IDCarga',
    how='left'
)
```

```python
df_atracacao = df_atracacao.merge(
    df_Carga_Filtrado[['IDCarga', 'Origem']],
    on='IDCarga',
    how='left'
)
```

```python
df_atracacao = df_atracacao.merge(
    df_Carga_Filtrado[['IDCarga', 'VLPesoCargaBruta']],
    on='IDCarga',
    how='left'
)
```

```python
df_atracacao.info()
```

```python
df_atracacao.tail(5)
```

```python

df_atracacao['VLPesoCargaBruta'] = pd.to_numeric(df_atracacao['VLPesoCargaBruta'], errors='coerce')
```

```python
peso_por_porto = df_atracacao.groupby('Porto Atracação')['VLPesoCargaBruta'].sum().sort_values(ascending=False)

peso_por_porto.head(10)
```

```python
quantidade_por_porto = df_atracacao.groupby('Porto Atracação')['IDCarga'].count().sort_values(ascending=False)
quantidade_por_porto.head(10)
```
